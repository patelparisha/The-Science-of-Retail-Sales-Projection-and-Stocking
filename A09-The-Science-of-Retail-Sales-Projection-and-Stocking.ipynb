##1. Dataset Overview

### **Amazon Sale Report Dataset**
Amazon Sale Report data includes the purchase history of individual customers and order details. It records such crucial components as order numbers, products, prices, as well as quantities, shipping, courier status, and type of customer. This information is the base of the transactional analysis as it demonstrates what customers bought, when they bought it and according to what conditions of fulfillment the orders were executed. It allows identifying the best-selling categories, assessing the B2B and B2C relations, identifying the promotion-based behavior and determining the purchasing trends of the regions in India.

### **Sale Report Dataset**
The Sale Report data is the operational counterpart of the transactional data. It holds SKU-level product data such as styles, sizes, category classifications, stock available and metadata on fulfillment. It provides a finer grain perspective of the product line, which will enable us to find out how the factors of size availability, style families, and operational settings affect the cancellation rates, sales rates, and delivery-driven performance of various product lines.

### **Integration**
These two datasets combined form one comprehensive structure which incorporates transactions behavior and details. The Amazon Sale Report gives the information about the what, when, and how many of the customer purchases, whereas the data of the Sale Report gives the information about the context under the SKU level which causes the outcomes. With their integration, it is possible to have an in-depth analysis of revenue drivers, category concentration, efficiency of channels of fulfillment and regional demand variation.

```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
-- Create a fully cleaned Amazon Sale Report table
-- Includes city/state standardization, courier status fix, and SKU cleanup


CREATE OR REPLACE TABLE `ba775-fall2025-a09.data.cleaned_amazon_sale_report` AS

WITH base_data AS (
  -- Step 1: Start from the raw Amazon sale report
  SELECT
    `Order ID`,
    `Date`,
    `Status`,
    `Fulfilment`,
    `Sales Channel`,
    `ship service level`,
    `Style`,
    `SKU`,
    `Category`,
    `Size`,
    `Courier Status`,
    `Qty`,
    `Amount`,
    `ship city` AS `Ship City`,
    `ship state` AS `Ship State`,
    `ship postal code` AS `Ship Postal Code`,
    `promotion ids`,
    `B2B`,
    `fulfilled by`
  FROM `ba775-fall2025-a09.data.amazon_sale_report`
),

postal_city_state_map AS (
  -- Step 2: Build mapping of Postal Code → standardized City + State
  SELECT
    `Ship Postal Code`,
    ANY_VALUE(TRIM(UPPER(`Ship City`))) AS `Ship City`,
    ANY_VALUE(TRIM(UPPER(`Ship State`))) AS `Ship State`
  FROM base_data
  WHERE `Ship City` IS NOT NULL AND `Ship State` IS NOT NULL
  GROUP BY `Ship Postal Code`
),

cleaned_base AS (
  -- Step 3: Join postal map + fix courier status for cancelled orders
  SELECT
    b.* EXCEPT(`Ship City`, `Ship State`, `Courier Status`),
    pcm.`Ship City` AS `Ship City`,
    pcm.`Ship State` AS `Ship State`,
    CASE
      WHEN b.`Courier Status` IS NULL AND b.`Status` = 'Cancelled' THEN 'Cancelled'
      ELSE b.`Courier Status`
    END AS `Courier Status`
  FROM base_data AS b
  LEFT JOIN postal_city_state_map AS pcm
    ON b.`Ship Postal Code` = pcm.`Ship Postal Code`
),

sku_clean AS (
  -- Step 4: Clean SKU (keep only first and last parts)
  SELECT
    b.`Order ID`,
    b.`Date`,
    b.`Status`,
    b.`Fulfilment`,
    b.`Sales Channel`,
    b.`ship service level`,
    b.`Style`,
    b.`Category`,
    b.`Size`,
    b.`Courier Status`,
    b.`Qty`,
    b.`Amount`,
    b.`Ship City`,
    b.`Ship State`,
    b.`Ship Postal Code`,
    b.`promotion ids`,
    b.`B2B`,
    b.`fulfilled by`,
    CONCAT(
      SPLIT(b.SKU, '-')[OFFSET(0)],
      '-',
      SPLIT(b.SKU, '-')[OFFSET(ARRAY_LENGTH(SPLIT(b.SKU, '-')) - 1)]
    ) AS SKU
  FROM cleaned_base AS b
)

-- Step 5: Final cleaned dataset output
SELECT *
FROM sku_clean;
""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

### 1.3: To what extent is there a relationship between an item’s price and the likelihood of it being cancelled or returned?

```python
from google.cloud import bigquery

client = bigquery.Client(project="ba775-fall2025-a09")

query = """
WITH sku_status AS (
 SELECT
   SKU,
   AVG(Amount) AS avg_price,
   COUNT(*) AS total_orders,
   COUNTIF(Status = 'Cancelled') AS cancelled_orders,
   COUNTIF(Status LIKE 'Shipped - Returned%') AS returned_orders
 FROM `ba775-fall2025-a09.data.cleaned_amazon_sale_report`
 GROUP BY SKU
)
SELECT
 SKU,
 avg_price,
 total_orders,
 cancelled_orders,
 returned_orders,
 SAFE_DIVIDE(cancelled_orders, total_orders) AS cancellation_rate,
 SAFE_DIVIDE(returned_orders, total_orders) AS return_rate
FROM sku_status
WHERE total_orders >= 5
ORDER BY cancellation_rate DESC;
"""

df = client.query(query).to_dataframe()
df
```

Cancellation rates remain relatively uniform across price levels, with high-cancellation SKUs falling within the same price range as zero-cancellation SKUs, indicating that item price is not a primary driver of cancellation behavior. The highest cancellation rate is 87.5% with an avg price of 487 INR and second highest of 83.3% has an avg price of 987.

**Limitation:** This analysis does not control for confounding factors such as fulfillment method, shipping speed, or category, which may independently influence cancellations.

### 2.2: Do promotional orders generate higher or lower average order value (AOV)?

```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
SELECT
CASE
  WHEN `promotion ids` IS NOT NULL THEN 'With Promotion'
  ELSE 'Without Promotion'
END AS promotion_status,
COUNT(*) AS total_orders,
ROUND(AVG(Amount), 2) AS avg_revenue_per_order,
ROUND(SUM(Amount), 2) AS total_revenue
FROM `ba775-fall2025-a09.data.cleaned_amazon_sale_report`
GROUP BY promotion_status;

""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

Promotional orders have a higher AOV of ₹674 vs ₹599 for non-promo (+12.5%). They also generate 2× total revenue due to higher volume. Promotions increase both per-order spending and total sales volume.

**Limitation:** Does not isolate whether discounts cause higher value or attract inherently higher-spending customers.

Yes. B2B orders have higher AOV (₹701) compared to non-B2B (₹648, +8.2%). But B2B has only 871 orders, making it a high-value but low-frequency segment.

**Limitation:** Small volume limits statistical confidence.
